```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Định nghĩa hàm mất mát giả lập (Hàm lồi)
def loss_function(w1, w2):
    return w1**2 + w2**2 + 5

# Tạo lưới tọa độ
w1 = np.linspace(-10, 10, 100)
w2 = np.linspace(-10, 10, 100)
W1, W2 = np.meshgrid(w1, w2)
Z = loss_function(W1, W2)

# Giả lập quá trình Gradient Descent
path_w1, path_w2 = [8.5], [7.0]
lr = 0.15
for _ in range(12):
    curr_w1, curr_w2 = path_w1[-1], path_w2[-1]
    path_w1.append(curr_w1 - lr * 2 * curr_w1)
    path_w2.append(curr_w2 - lr * 2 * curr_w2)
path_z = loss_function(np.array(path_w1), np.array(path_w2))

# Vẽ đồ thị 3D
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(W1, W2, Z, cmap='terrain', alpha=0.6)
ax.plot(path_w1, path_w2, path_z, color='red', marker='o', label='Đường hội tụ')
ax.set_xlabel('Trọng số w1'); ax.set_ylabel('Trọng số w2'); ax.set_zlabel('Loss L(w)')
plt.legend(); plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0.001, 1, 100)
y = -np.log(x)

plt.figure(figsize=(8, 5))
plt.plot(x, y, label=r'$L = -\log(\hat{y})$', color='red', linewidth=2)
plt.fill_between(x, y, color='red', alpha=0.1)
plt.title('Cơ chế phạt của hàm mất mát Cross-Entropy')
plt.xlabel('Xác suất dự đoán cho lớp đúng ($\hat{y}$)')
plt.ylabel('Giá trị mất mát (Loss)')
plt.grid(True, linestyle='--')
plt.legend()
plt.show()
```
